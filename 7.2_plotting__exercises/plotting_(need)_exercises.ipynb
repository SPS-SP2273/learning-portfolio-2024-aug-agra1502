```python
from matplotlib import pyplot as plt
```

## Exericse 1

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=0.25, yerr=y_err,
             color='r', linestyle='dashed',
             marker='o', markerfacecolor='y',
             ecolor=[0, 0, 0],capsize=5,
             label='X vs Y')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(['Y value with errors'], loc='lower right')
plt.show()
```

## Exercise 2

```python
import numpy as np

# Some data for plotting
x = np.arange(0,50)
x = np.linspace(0, 50, 300)
y = np.exp(-x/10)*np.sin(x)

plt.errorbar(x, y, 
             color='b',
             ecolor=[0, 0, 0],capsize=5,
             label='X vs Y')
plt.xlabel('x', fontsize=15, color='xkcd:dark orange')
plt.ylabel('y',fontsize=15, color='xkcd:dark orange', rotation=0)
plt.ylim(-1, 1)
plt.title(r'$y=e^{\frac{-x}{10}}\mathrm{sin}(x)$', fontsize=20)
plt.grid(alpha=.25)
plt.show()
```
